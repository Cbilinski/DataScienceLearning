```python
import pandas as pd
import ipywidgets as widgets
import numpy as np
import matplotlib.pyplot as plt
import sklearn.linear_model as lm 
!pip install --upgrade @jupyter-widgets/controls
```

```python
df = pd.read_csv("Data/marketing.csv")
df
```

```python
def plot_graph(income=0, amt_spent=0):
    x_values = np.linspace(0, 160000, 160000)
    # run linear regression on Income vs. Meat Purchase information and plot it
    x = df[["Income"]]
    y = df["MntMeatProducts"]
    lin_reg = lm.LinearRegression()
    lin_reg.fit(x, y) 
    plt.scatter(x, y)
    ypred = lin_reg.predict(x)
    plt.plot(x, ypred, color="purple")

    # run linear regression on Income vs. Fish Purchase information and plot it
    x = df[["Income"]]
    y = df["MntFishProducts"]
    lin_reg = lm.LinearRegression()
    lin_reg.fit(x, y) 
    plt.scatter(x, y)
    ypred = lin_reg.predict(x)
    plt.plot(x, ypred, color="red")

    # add plot labels
    plt.xlabel("Income")
    plt.ylabel("Amount spent on meat or fish")
    plt.title("Best fit lines for customers income vs. spending habits")
    plt.ylim([0,1000])
    plt.plot(x_values, 0*x_values + amt_spent, color='black')
    plt.axvline(x=income, color='black')

# Manually input income and amt_spent values
income = int(input("Enter income value (between 0 and 160000): "))
amt_spent = int(input("Enter amount spent value (between 10 and 1000): "))

# Call the plot_graph function with the input values
plot_graph(income, amt_spent)
#widgets.interact(plot_graph, income=(0,160000,5000), amt_spent=(10, 1000, 10) )
```

## Dataset

The code cells above load a csv file containing information about the spending habits of customers based on their income. 

Run all the cells in order to produce a plot of Income on the x-axis, and amount spent on meat (blue dots) or fish (orange dots) on the y-axis. 

The plot also includes the result of running linear regression on the Income vs. Meat Spending (purple line) and Income vs. Fish spending (red line). 

The plot is interactive. 
* You can move the slider labeled `income` to move a black vertical line to different income values. 
* You can move the slider labeled `amt_spent` to move a black horizontal line to different amount spent values. 

## Question 1

Using the sliders and the linear regression results, determine the average amount that a customer with an income of 80,000 spends on meat?

Using the sliders and the linear regression results, determine the average amount that a customer with an income of 80,000 spends on fish?


```python
# Extracting the linear regression coefficients for Meat Purchase
meat_lin_reg = lm.LinearRegression()
x_meat = df[["Income"]]
y_meat = df["MntMeatProducts"]
meat_lin_reg.fit(x_meat, y_meat)
meat_coef = meat_lin_reg.coef_[0]
meat_intercept = meat_lin_reg.intercept_

# Predicting the amount spent on meat for an income of $80,000
income_80000 = 80000
predicted_amount_meat_80000 = meat_coef * income_80000 + meat_intercept

# Rounding the predicted amount to the nearest whole number
rounded_predicted_amount_meat_80000 = round(predicted_amount_meat_80000)

# Printing the result
print("The average predicted amount that a customer with an income of $80,000 spends on meat is:", rounded_predicted_amount_meat_80000)
```

```python
# Extracting the linear regression coefficients for Fish Purchase
fish_lin_reg = lm.LinearRegression()
x_fish = df[["Income"]]
y_fish = df["MntFishProducts"]
fish_lin_reg.fit(x_fish, y_fish)
fish_coef = fish_lin_reg.coef_[0]
fish_intercept = fish_lin_reg.intercept_

# Predicting the amount spent on fish for an income of $80,000
income_80000 = 80000
predicted_amount_fish_80000 = fish_coef * income_80000 + fish_intercept

# Rounding the predicted amount to the nearest whole number
rounded_predicted_amount_fish_80000 = round(predicted_amount_fish_80000)

# Printing the result
print("The average predicted amount that a customer with an income of $80,000 spends on fish is:", rounded_predicted_amount_fish_80000)
```
