## MEAN AND STANDARD DEVIATION

```python
pd.set_option('display.max_columns', 500)
df = pd.read_csv("Data/instagram_followers.csv")
df
```

## Question 1

What is the mean number of followers for all accounts under the "Sports" category? And what is the mean number of followers for all accounts under the "Entertainment" category? 

```python
# Calculate the mean number of followers for each category
sports_mean = df[df['Category'] == 'Sports']['Followers'].mean()
entertainment_mean = df[df['Category'] == 'Entertainment']['Followers'].mean()
print("Mean number of followers for Sports category:", sports_mean)
print("Mean number of followers for Entertainment category:", entertainment_mean)
```

## Question 2

What is the standard deviation of the number of followers for all accounts under the "Sports" category? And what is the standard deviation of the number of followers for all accounts under the "Entertainment" category? 

```python
# Calculate the mean number of followers for each category
sports_std = df[df['Category'] == 'Sports']['Followers'].std()
entertainment_std = df[df['Category'] == 'Entertainment']['Followers'].std()
print("STD of followers for Sports category:", sports_std)
print("STD of followers for Entertainment category:", entertainment_std)
```
